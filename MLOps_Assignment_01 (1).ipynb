## Explain Context window of LLMs in a few lines and why this metric is important

**(context)**
the information user gives to LLM when asking something  
e.g: user's question to chat gpt like explain E=mc^2 to 7 yeals old student

**(context window)**
max size of content
e.g: max size of words / tokens we ask to chat gpt

**importance of metrics**
related with performance,  large size results in better relevant response at cost of computational power. low size is faster but may result in repetition and low relevancy
  

# Explain the self attention mechanism in LLMs


in simple words, self attention in LLM is the focusing and understanding different parts of sentense in a sequence in parallel in order to better understand the meaning / context

**how its implemented**
by creating 3 vectors i.e: query, key and value
query = word / token being asked followed by sequence
key = vector that is rerference of q
value = vector that shows meaning / answer

**similarity score** is calculated via query / key
**weighted sum** of value vector is calculated, higer score means better. afrer encoding this sum its used to predicting next word / answer

# Explain what is meant by Quantization of a LLM


Quantization is very good technique for making our LLM more effective provided the balancing is optimally maintained with precision by lowering the resolution (memory size / representation)
typically we convert from higher resolution to lower, like from FP32 to int8
it can be done

consider an example of a 32bit PNG with 24MP CMYK or RGB colours, if such photo is fed for training as is it will eat up GPU resulting in more cost,  if photograph is general like animal, Football then with neglectable losses with training it can  output better results even if photo is made from 8bpp and even low dimention,  here quantization only deals with bpp (bits per pixel)
